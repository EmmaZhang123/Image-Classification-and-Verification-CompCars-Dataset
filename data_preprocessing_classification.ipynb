# Data Preprocessing For Classification Task

```python
import os
from PIL import Image
import shutil
import pandas as pd
```

## Data Split

```python
'''We choose the dataset from the subset for calssification in Paper'''

# Define paths
base_dir = 'dataset/data'
image_dir = os.path.join(base_dir, 'crop_image')
train_list_file = os.path.join(base_dir, 'train_test_split/classification/train.txt')
test_list_file = os.path.join(base_dir, 'train_test_split/classification/test.txt')
train_output_dir = 'output/train'
test_output_dir = 'output/test'
train_labels_file = 'output/train_labels.csv'
test_labels_file = 'output/test_labels.csv'

# Ensure output directories exist
os.makedirs(train_output_dir, exist_ok=True)
os.makedirs(test_output_dir, exist_ok=True)

# Function to copy images and save labels
def copy_images_and_save_labels(list_file, output_dir, labels_file):
    data = []
    with open(list_file, 'r') as f:
        lines = f.readlines()
        for line in lines:
            image_path = line.strip()
            make_id = image_path.split('/')[0]  # Extract the make_id
            image_name = os.path.basename(image_path)
            src_path = os.path.join(image_dir, image_path)
            dst_path = os.path.join(output_dir, image_name)
            shutil.copy2(src_path, dst_path)
            data.append({'image': image_name, 'label': make_id})
            print(f'Copied {src_path} to {dst_path} with label {make_id}')
    # Save labels to CSV
    df = pd.DataFrame(data)
    df.to_csv(labels_file, index=False)
    print(f'Labels saved to {labels_file}')

# Copy train images and save labels
copy_images_and_save_labels(train_list_file, train_output_dir, train_labels_file)

# Copy test images and save labels
copy_images_and_save_labels(test_list_file, test_output_dir, test_labels_file)

print('All images have been copied and labels have been saved successfully.')
```

## Crop the Original Images

```python
# Define paths
image_base_dir = 'dataset/data/image'
label_base_dir = 'dataset/data/label'
output_base_dir = 'dataset/data/crop_image'

# Ensure output directory exists
os.makedirs(output_base_dir, exist_ok=True)

def process_image(image_path, label_path, output_path):
    try:
        # Open image
        with Image.open(image_path) as img:
            # Read label file
            with open(label_path, 'r') as label_file:
                lines = label_file.readlines()
                
                # Viewpoint annotation (can be used if needed)
                viewpoint = int(lines[0].strip())
                
                # Number of bounding boxes
                num_boxes = int(lines[1].strip())
                
                if num_boxes == 1:
                    # Bounding box coordinates
                    bbox_coords = lines[2].strip().split()
                    x1, y1, x2, y2 = map(int, bbox_coords)
                    
                    # Crop image using bounding box
                    cropped_img = img.crop((x1, y1, x2, y2))
                    
                    # Save cropped image
                    cropped_img.save(output_path)
                else:
                    print(f"Unexpected number of bounding boxes: {num_boxes}")
    except Exception as e:
        print(f"Error processing {image_path}: {e}")

def process_all_images(image_base_dir, label_base_dir, output_base_dir):
    for root, dirs, files in os.walk(image_base_dir):
        for file in files:
            if file.endswith('.jpg'):
                # Construct paths
                image_path = os.path.join(root, file)
                rel_path = os.path.relpath(root, image_base_dir)
                label_file = file.replace('.jpg', '.txt')
                label_path = os.path.join(label_base_dir, rel_path, label_file)
                
                if not os.path.exists(label_path):
                    print(f"Label file not found: {label_path}")
                    continue
                
                # Define output path maintaining the same subdirectory structure
                output_subdir = os.path.join(output_base_dir, rel_path)
                os.makedirs(output_subdir, exist_ok=True)
                output_path = os.path.join(output_subdir, file)
                
                # Process image
                process_image(image_path, label_path, output_path)

# Run the process
process_all_images(image_base_dir, label_base_dir, output_base_dir)
```

```python
'''Change the original label to 0-74'''
# Define the original and new labels
original_labels = [78, 81, 77, 54, 100, 102, 105, 48, 58, 82, 76, 150, 152,
                   73, 4, 36, 35, 34, 122, 134, 121, 87, 71, 11, 92, 53,
                   160, 155, 15, 97, 95, 68, 80, 118, 131, 45, 27, 50, 142,
                   51, 140, 132, 94, 29, 158, 106, 39, 138, 88, 114, 64, 69,
                   108, 148, 159, 55, 133, 162, 75, 120, 26, 111, 89, 149, 119,
                   14, 103, 70, 28, 128, 43, 157, 38, 84, 37]
new_labels = list(range(0, 75))
label_mapping = dict(zip(original_labels, new_labels))

# Load the CSV files
train_labels = pd.read_csv('dataset_classification/train_labels.csv')
test_labels = pd.read_csv('dataset_classification/test_labels.csv')

# Apply the mapping to the labels
train_labels['label'] = train_labels['label'].map(label_mapping)
test_labels['label'] = test_labels['label'].map(label_mapping)

# Save the updated CSV files
train_labels.to_csv('train_labels_mapped.csv', index=False)
test_labels.to_csv('test_labels_mapped.csv', index=False)
```
